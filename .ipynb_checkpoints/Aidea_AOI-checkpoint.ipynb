```python
from keras.applications.inception_v3 import InceptionV3
from keras.layers import Conv2D,MaxPooling2D,Activation,Flatten,Dense,Dropout
from time import gmtime, strftime
import matplotlib.pyplot as plt
```

改善過擬合的其中一個方法是增加資料，可以用keras自帶的一個資料增強Data Augmentation

```python
train_datagen = ImageDataGenerator(
      rescale=1./255, #影象張量的數字縮放
      rotation_range=40, # 角度值，0~180，影象旋轉
      width_shift_range=0.2, # 水平平移，相對總寬度的比例
      height_shift_range=0.2, # 垂直平移，相對總高度的比例
      shear_range=0.2, # 隨機錯切換角度
      zoom_range=0.2, # 隨機縮放範圍
      horizontal_flip=True, # 一半影象水平翻轉
      fill_mode='nearest') # 填充新建立畫素的方法
```

```python
train_generator = train_datagen.flow_from_directory(
      train_path, # 目標目錄
      target_size=(224, 224), # 所有影象調整為150x150
      classes=['0', '1', '2', '3', '4', '5'],
      batch_size=32,
      class_mode='categorical')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)

validation_generator = test_datagen.flow_from_directory(
        valid_path,
        target_size=(224, 224),
        classes=['0', '1', '2', '3', '4', '5'],
        batch_size=32,
        class_mode='categorical')
```

```python
print(validation_generator)
```

## 建立模型

```python
conv_base = VGG16(weights='imagenet', 
                  include_top=False,
                  input_shape=(224,224,3))

model = models.Sequential()
model.add(conv_base)
```

```python
model.add(Conv2D(32, (1, 1),activation="relu"))
model.add(Dropout(0.2))
model.add(Conv2D(64,(3, 3),activation="relu"))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(32))
model.add(Dense(64))
model.add(Dense(64))

model.add(Dense(6,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.RMSprop(lr=2e-5),
              metrics=['acc'])
```

```python
model.summary()
```

```python
print(train_generator.samples//train_generator.batch_size,)
print(validation_generator.samples//validation_generator.batch_size)
```

```python
history = model.fit_generator(
      train_generator,
      steps_per_epoch=50,
      epochs=15,
      validation_data=validation_generator,
      validation_steps=10
)
```

```python
test_generator = test_datagen.flow_from_directory(
        test_path,
        target_size=(224, 224),
        classes=['0', '1', '2', '3', '4', '5'],
        batch_size=32,
        class_mode='categorical')
```

```python
test_loss, test_acc = model.evaluate_generator(
	test_generator, 
	steps=test_generator.samples//test_generator.batch_size)
```

```python
print('test acc:', test_acc)
```
